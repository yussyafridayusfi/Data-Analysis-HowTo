```python
import pandas as pd
import numpy as np
```

## Read All Dataset
### Sources Kaggle : https://www.kaggle.com/datasets/shivamb/netflix-shows
Amazon Prime, Disney Plus, Hulu, Netflix

```python
df_amazon = pd.read_csv("../dataset/amazon_prime_titles.csv")
df_amazon.head(3)
```

```python
df_disney = pd.read_csv("../dataset/disney_plus_titles.csv")
df_disney.head(3)
```

```python
df_hulu = pd.read_csv("../dataset/hulu_titles.csv")
df_hulu.head(3)
```

```python
df_netflix = pd.read_csv("../dataset/netflix_titles.csv")
df_netflix.head(3)
```

### Create a new column for source of each dataset
df_name[column_name] = column_value

```python
df_amazon["source"] = "Amazon Prime"
df_amazon.head(3)
```

```python
df_hulu["source"] = "Hulu"
df_hulu.head(3)
```

```python
df_disney["source"] = "Disney Plus"
df_disney.head(3)
```

```python
df_netflix["source"] = "Netflix"
df_netflix.head(3)
```

### Merge all dataframe into one using concat
pd.concat([df_name])

```python
union_df = pd.concat([df_amazon,df_hulu,df_disney,df_netflix])
union_df.head()
```

### Describe the new dataset after merge
Check count, describe, values, show random sample of data.

```python
union_df.count()
```

```python
df_amazon.count()
```

```python
union_df.sample(20)
```

### Create a dataset
Create new dataset from merge multiple dataset into a csv file.

```python
union_df.to_csv("../dataset/dataset_join.csv", sep=',', index = False, encoding='utf-8')
```
